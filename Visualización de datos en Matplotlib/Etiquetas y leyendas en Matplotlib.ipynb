```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/usuario/Desktop/archivos_base_python_data_science_big_data_esencial/basedatos2008.csv", nrows=100000)
#creamos una tabla resumen que son el numero de vuelos que salieron los dias de semana
data = np.unique(df.DayOfWeek,return_counts=True) #variable que contendra la columna DayOfWeek que tiene valores del 1 a 7 y ademas revolvera el conteo de cada uno de esos valores
labs = ["Lun","Mar","Mie","Jue","Vie","Sab","Dom"] #seran las etiquetas de nustros datos
#data
```

```python
#grafico de sectores
plt.pie(x=data[1], # introducira los valores del nivel 1 del Array 
       labels=data[0], # introducira los valores del nivel 0 del Array 
       radius=1.5,
       colors=["Red","Green","Orange","Blue","Gray","Pink","Black"],
       startangle=90) # hace que en el grafico se empiece a contar a partir del angulo 90
plt.show()
```

```python
#añadimos reemplazamos labels por labs que habiamos creado
plt.pie(x=data[1],
       labels=labs,
       radius=1.5,
       colors=["Red","Green","Orange","Blue","Gray","Pink","Black"],
       startangle=90)
plt.show()
```

```python
#utilizamos paletas de la libreria seaborn (sns)
plt.pie(x=data[1],
       labels=labs,
       radius=1.5,
       colors=sns.color_palette("hls",7),
       startangle=90)
plt.show()
```

```python
#añadimos el objeto explode (0.1 = .1)
plt.pie(x=data[1],
       labels=labs,
       radius=1.5,
       colors=sns.color_palette("hls",7),
        explode= (0,0,0,0,0,0,0.1),
       startangle=90)
plt.show()
```

```python
#añadimos porcentaje autopct
plt.pie(x=data[1],
       labels=labs,
       radius=1.5,
       colors=sns.color_palette("hls",7),
        explode= (0,0,0,0,0,0,0.1),
       startangle=90,
       autopct="%1.1f%%" )
plt.show()
```

```python
#añadimos distancia de label
plt.pie(x=data[1],
       labels=labs,
       radius=1.5,
       colors=sns.color_palette("hls",7),
        explode= (0,0,0,0,0,0,0.1),
       startangle=90,
       autopct="%1.1f%%",
       labeldistance = 1.1)
plt.show()
```

```python
#añadimos distancia de label
plt.pie(x=data[1],
       #labels=labs,
       radius=0.5,
       colors=sns.color_palette("hls",7),
        explode= (0,0,0,0.2,0,0,0.1),
       startangle=90,
       autopct="%1.1f%%")
plt.legend(loc="upper left",labels=labs)
plt.show()
```

```python
#Grafico de barras con la libreria seaborn (sns)
plt = sns.barplot(x=labs, y=data[1])
plt.set(xlabel="Día de la Semana", ylabel="números de vuelos")
```
